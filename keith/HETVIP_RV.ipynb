```python
def fill_nan_21(wave, fluxes,err,clean_outlier=None,RV=None,chop_edge = None):
    from tqdm import tqdm
    import numpy as np
    from scipy import interpolate
    import astropy.stats
    fluxes= array(fluxes)
    #inds = np.arange(fluxes.shape[0])
    
    if chop_edge is not None:
        wave = wave[chop_edge:-chop_edge]
        F = [] ; E = []
        for i in range(len(fluxes)):
            F.append(fluxes[i][chop_edge:-chop_edge])
            E.append(err[i][chop_edge:-chop_edge])
        fluxes= array(F)
        err = array(E)
        
    newspec = []
    for i in tqdm(range(len(fluxes ))):

        if clean_outlier is not None:
            fluxes[i] = astropy.stats.sigma_clip(fluxes[i],sigma_upper=clean_outlier,sigma_lower=10,maxiters=2)
            #fluxes[i]=fluxes[i].filled(fill_value=nan)
        #if RV is not None:
        #	factor = RV[i] / 2.99792458E5
        #	waves = sqrt((1+factor)/(1-factor))*waves

        #good = np.where(np.isfinite(fluxes[i]))[0]
        good = where(isnan(fluxes[i]) == False)[0]
        bad = where(isnan(fluxes[i]) == True)[0]
        err[i][bad] =  nanmedian(err[i]) * 1000 
        #wavelim=[3600,5400]
        #wavenew = linspace(min(wave),max(wave),num=len(fluxes[i]) )
        wavenew = wave

        #print shape(waves), shape(good),shape(fluxes[i]), shape(waves[good]), shape(fluxes[good])
        #print good, fluxes[i][good],i 
        f = interpolate.interp1d(wave[good], fluxes[i][good],"cubic", fill_value="extrapolate")
        newspec.append(f(wavenew))
    wavenew = [linspace(min(wave),max(wave),num=len(fluxes[i]) )] * len(fluxes)
    wavenew = array(wavenew) ; newspec=array(newspec)
    p.plot(wavenew,newspec/median(newspec))
    #B = np.where(np.isfinite(A),A,f(inds))
    return newspec,err



def BMCcntflux(lams, fluxes,width = 50,ptile=98, cpy = False,plt=False):
    import copy as cp
    import time
    from scipy import stats
    continuumflux = []

    for i in arange(len(fluxes)):
        testind = where((lams >= lams[i]-(width/2.0)) & (lams <= lams[i]+(width/2.0)))[0]
        continuumflux.append(stats.scoreatpercentile(fluxes[testind],ptile))

    if plt:
        p.plot(lams,fluxes)
        p.plot(lams,continuumflux,'-')
    fluxes /= continuumflux
    return fluxes

def BMCcntflux_poly(lams, fluxes,width = 20,ptile=98, cpy = False,plt=False,deg = 3):
    import numpy as np
    import copy as cp
    continuumpts = []
    lampts = []

    for i in arange(len(fluxes)):
        testind = where((lams >= lams[i]-(width/2.0)) & (lams <= lams[i]+(width/2.0)))[0]
        lampts.append( (lams[i]-(width/2.0) +lams[i]+(width/2.0) )/2.0 )
        if isnan(percentile(fluxes[testind],ptile) ) == False:
            continuumpts.append(percentile(fluxes[testind],ptile))
        else:
            continue
    try:
        Z  = np.polyfit(lampts, continuumpts,deg=deg)
    except TypeError:
        print('WARNING couldnt fit conitnuum returning nans')
        return np.array([nan]*len(lams))
    func = np.poly1d(Z)
    continuumflux = func(lams)
    if plt:
        p.plot(lams,fluxes)
        p.plot(lams,continuumflux,'-')
    fluxes /= continuumflux
    return fluxes
```

```python
def test_CCF(obs_spc, template_spec,template_labels):
    #template_spec, template_labels = grid_load(grid_path)
    okind = where(isnan(obs_spc) == False)[0]
    obs_spc = obs_spc[okind]
    template_spec = array(template_spec).T[okind].T


    template_spec = np.atleast_2d(template_spec)
    N = len(obs_spc)
    N = N - 1 if N % 2 > 0 else N
    obs_spec = obs_spc[:N]
    template_spec = template_spec[:,:N]
    N_templates = len(template_spec)
    try:
        fft_obs_flux = np.fft.fft(obs_spec)
        fft_template_flux = np.fft.fft(template_spec)
        template_flux_corr = (fft_obs_flux * fft_template_flux.conjugate())
        template_flux_corr /= np.sqrt(np.inner(obs_spec, obs_spec))
        R = np.ones(N_templates) * np.nan
    except:
        return 'NAN'
    for i in range(N_templates):
        denominator = np.sqrt(np.inner(template_spec[i], template_spec[i]))
        flux_correlation = template_flux_corr[i] / denominator
        correlation = np.fft.ifft(flux_correlation).real
        ccf = np.zeros(N)
        ccf[:N//2] = correlation[N//2:]
        ccf[N//2:] = correlation[:N//2]
        h = ccf.max()
        R[i] = h
    
    try :
        return template_labels[np.nanargmax(R)]
    except :
        return 'NAN'

def test_CCF_all(D, template_spec, template_names, weight_lim=0.6, M=None):
    from tqdm import tqdm
    #ok = where((D['bp_rp'] >= 0.8)&(D['bp_rp'] < 2.0))[0]
    #print('There are %i stars'%(len(ok)))
    #D=D[ok]
    temp = []
    temps = [] 
    for i in tqdm(range(len(D))):
        #indi = where(BB['source_id'] == B['source_id'][i])[0]
        #if len(indi) >= 1:
        #	temps.append(BB['RV_temp_v1'][indi[0]])
        #	continue
        T = test_CCF(D['norm_flux'][i],template_spec,template_names )
        temps.append(T)
    D['RV_temp_v1'] = temps
    return D
```

```python
def grid_load(grid_path = '/Users/khawkins/Box/RVgrid/',typ = None,width=50,ptile=98,):
    from scipy.interpolate import interp1d
    D = loadtxt(grid_path + 'allspecs_orig.txt',dtype=str)
    specs = []
    labs = []
    datawave = linspace(3470,5540,num=1036)
    for i in range(len(D)):

        #base = D[i].split('_')
        #T = int(base[0].split('T')[1])
        #g = float(base[1].split('g')[1])
        #f = float(base[2].split('f')[1].split('.')[0])
        #labs.append([T,g,f])
        labs.append(D[i])
        S = loadtxt(grid_path+D[i])
        
        f2 = interp1d(S[:,0], S[:,1], kind='cubic')
        f3 = interp1d(S[:,0], S[:,2], kind='cubic')
        if typ is None:
            SS = S[:,1]
            f2 = interp1d(S[:,0], SS, kind='cubic')
        elif typ == 'BMC':
            SS = BMCcntflux(S[:,0], S[:,1],width = width,ptile=ptile, cpy = False,plt=False)


        elif typ =='BMCpoly':
            SS = BMCcntflux_poly(S[:,0], S[:,1],width = width,ptile=ptile, cpy = False,plt=False,deg=3)
            name = D[i].split('.spec.conv')[0] + '_BMCpoly.spec'
            print(name)
            f = interp1d(S[:,0], SS, kind='cubic')
            wrt = array([datawave,f2(datawave),f3(datawave),f(datawave)]).T
            savetxt(grid_path + name, wrt, fmt='%.3f',delimiter='\t')
            specs.append(f(datawave))
        else:
            raise ValueError('typ needs to be None, BMC, or BMCpoly')
        f = interp1d(S[:,0], SS, kind='cubic')
        specs.append(f(datawave))
    return labs, specs
```

```python
def determine_radial_velocity_iSpec(wave_temp,fluxes_temp,wave_obj,fluxes_obj,err_obj, vel_lim=[-500,500],\
                                    velstep=1,correct=False,RVtyp='temp',wavefilt=[3480,5380]):
    import os
    import copy
    import sys
    import numpy as np
    import logging
    import multiprocessing
    from multiprocessing import Pool
    import astropy.io.fits as pyfits
    ispec_dir = '/Users/khawkins/iSpec_v20201001/'
    sys.path.insert(0, os.path.abspath(ispec_dir))
    import ispec
    temppath = '/Users/khawkins/Desktop/data/HETDEX/RVgrid/'
    
    obj_spectrum = ispec.create_spectrum_structure(wave_obj/10.)
    obj_spectrum['flux'] =  fluxes_obj
    obj_spectrum['err'] = err_obj
    
    #wave_temp = /10.
    
    template = ispec.create_spectrum_structure(wave_temp/10.) 
    template['flux'] = fluxes_temp 
    template['err'] = ones(len(wave_temp))*0.01
    
    if RVtyp =='temp':
        print('HERE')
        models, ccf = ispec.cross_correlate_with_template(obj_spectrum, template, lower_velocity_limit=vel_lim[0],\
                                                          upper_velocity_limit=vel_lim[1], velocity_step=velstep, fourier=True)
        components = len(models)
        try:
            rv = np.round(models[0].mu(), 2) # km/s
            rv_err = np.round(models[0].emu(), 2)
            rv = list(np.atleast_1d(rv)) 
            rv_err = list(np.atleast_1d(rv_err))
        except:
            print('No RV can be found')
            return [nan], [nan],nan
        if components > 1:
            print('WARNING: Star has multiple RV components')
            for i in np.arange(components-1):
                rv.append(np.round(models[i+1].mu(), 2))
                rv_err.append(np.round(models[i+1].emu(), 2))
    if correct:
        print('Correcting %s using the first component of the RV = %.2f'%(spectra_file,rv[0]))
        corrected_spectrum = ispec.correct_velocity(obj_spectrum, rv[0])
        wfilter = ispec.create_wavelength_filter(corrected_spectrum, wave_base=wavefilt[0]/10., wave_top=wavefilt[1]/10.0)
        corrected_spectrum = corrected_spectrum[wfilter]
        corrected_spectrum['waveobs'] *=10.
        ispec.write_spectrum(corrected_spectrum, "%s"%correct)
    return rv, rv_err, components
```
